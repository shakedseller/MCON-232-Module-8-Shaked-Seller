# Question 1 — Recursive Countdown Sum

Create a recursive function:

```python
sum_to_n(n)
```

The function should recursively calculate the sum from:

```text
1 + 2 + 3 + ... + n
```

Example:

```python
sum_to_n(5)
```

returns:

```text
15
```

because:

```text
1 + 2 + 3 + 4 + 5 = 15
```

## Requirements

- Use recursion
- Create an appropriate base case
- Do not use loops
- Demonstrate your function with at least 3 test cases

````python
def sum_to_n(n):
    # TODO: create your base case
    if n == 1:
      return 1
    # TODO: create your recursive case
    value = sum_to_n(n - 1)
    return n + value

# Test cases
print(sum_to_n(5))   # expected: 15
print(sum_to_n(1))   # expected: 1
print(sum_to_n(10))  # expected: 55
````

# Question 2 — Recursive String Reverse

Create a recursive function:

```python
reverse_string(text)
```

The function should recursively reverse a string.

Example:

```python
reverse_string("hello")
```

returns:

```text
"olleh"
```

## Requirements

- Use recursion
- Do not use loops
- Do not use slicing shortcuts such as `[::-1]`
- Demonstrate your function with at least 3 test cases

````python
def reverse_string(text):
    # TODO: create your base case
    if len(text) == 1:
      return text
    # TODO: create your recursive case
    new_string = reverse_string(text[1:]) + text[0]
    return new_string


# Test cases
print(reverse_string("hello"))     # expected: olleh
print(reverse_string("recursion")) # expected: noisrucer
print(reverse_string("cat"))       # expected: tac
````

# Question 3 — Recursive Count of Vowels

Create a recursive function:

```python
count_vowels(text)
```

The function should recursively count the vowels in a string.

Example:

```python
count_vowels("recursion")
```

returns:

```text
4
```

## Requirements

- Count the vowels: `a`, `e`, `i`, `o`, `u`
- Uppercase vowels should also count
- Use recursion only
- Do not use loops
- Demonstrate your function with at least 3 test cases

````python
def count_vowels(text):
    vowels = {"a", "e", "i", "o", "u"}
    text = text.lower()
    #print(text)
    # TODO: create your base case
    if text == "":
      return 0
    # TODO: check whether the current character is a vowel
    # TODO: create your recursive case
    if text[0] in vowels:
      return 1 + count_vowels(text[1:])

    return count_vowels(text[1:])


# Test cases
print(count_vowels("recursion"))     # expected: 4
print(count_vowels("COMPUTER"))      # expected: 3
print(count_vowels("sky"))           # expected: 0
````

# Question 4 — Recursive Maximum Value

Create a recursive function:

```python
find_max(numbers, index)
```

The function should recursively find the largest value in a list.

Example:

```python
find_max([4, 9, 2, 11, 5], 4)
```

returns:

```text
11
```

## Requirements

- Use recursion
- Do not use Python's built-in `max()` function
- Use the `index` parameter to reduce the problem size
- Demonstrate your function with at least 3 test cases

````python
def find_max(numbers, index):
    # TODO: create your base case
    if index == 0:
      return numbers[index]
    # TODO: recursively find the maximum of the smaller list section
    # TODO: compare the current value with the recursive result
    if numbers[index] > find_max(numbers, index - 1):
      return numbers[index]
    else:
      return find_max(numbers, index - 1)
    index -= 1


# Test cases
print(find_max([4, 9, 2, 11, 5], 4))       # expected: 11
print(find_max([100, 3, 8, 12], 3))        # expected: 100
print(find_max([-10, -3, -25, -1], 3))     # expected: -1
````

# Question 5 — Recursive Binary Conversion

Create a recursive function:

```python
decimal_to_binary(n)
```

that converts a decimal integer into binary.

Example:

```python
decimal_to_binary(10)
```

returns:

```text
1010
```

---

# Binary Reference Table

```text
Decimal → Binary

0  →  0
1  →  1
2  →  10
3  →  11
4  →  100
5  →  101
6  →  110
7  →  111
8  →  1000
9  →  1001
```

---

# Important Recursive Hint

To convert a decimal number into binary recursively:

1. Divide the number by 2 using integer division:

```python
n // 2
```

2. The remainder becomes the current binary digit:

```python
n % 2
```

3. Continue recursively solving the smaller problem:

```python
n // 2
```

until you reach the base case(s).

---

# Example Walkthrough

Convert:

```text
10
```

Step-by-step:

```text
10 // 2 = 5   remainder = 0
5  // 2 = 2   remainder = 1
2  // 2 = 1   remainder = 0
1 is a base case
```

As recursion unwinds, the remainders combine to form:

```text
1010
```

---

# Requirements

- Use recursion
- Do not use:
  - loops
  - `bin()`
- Create appropriate base case(s)
- Demonstrate your function with at least 3 test cases

````python
def decimal_to_binary(n):
    # TODO: create your base case
    if n == 1 or n == 0:
      return str(n)
    # TODO: create your recursive case using division and remainder
    n_rem = n % 2
    return  decimal_to_binary(n // 2) + str(n_rem)

# Test cases
print(decimal_to_binary(10))  # expected: 1010
print(decimal_to_binary(7))   # expected: 111
print(decimal_to_binary(16))  # expected: 10000
````
